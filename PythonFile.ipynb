```python
import numpy as np
import pandas as pd
```

```python
movies=pd.read_csv('tmdb_5000_movies.csv')
credits=pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head(1)
```

```python
movies=movies.merge(credits,on='title')
```

```python
movies.head(1)
```

```python
movies=movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
import ast
ast.literal_eval  #convert the string into list
```

```python
def convert(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

```python
movies['genres']= movies['genres'].apply(convert)
```

```python
movies['keywords']= movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
def convert3(obj):
    l=[]
    c=0
    for i in ast.literal_eval(obj):
        if c!=3:
            l.append(i['name'])
            c+=1
        else:
            break
    return l
```

```python
movies['cast']= movies['cast'].apply(convert3)
```

```python
def fetch_director(obj):
    l=[]
    for i in ast.literal_eval(obj):
        if i['job']=="Director":
            l.append(i['name'])
            return l
```

```python
 movies['crew']= movies['crew'].apply(fetch_director)
```

```python
movies['overview']= movies['overview'].apply(lambda x:x.split())
```

```python
print(movies['crew'].to_string())
```

```python
movies['crew'] = movies['crew'].apply(lambda x: [i.replace(" ", '') for i in x] if x is not None else x)
```

```python
movies['genres']= movies['genres'].apply(lambda x:[i.replace(" ",'') for i in x])
movies['keywords']= movies['keywords'].apply(lambda x:[i.replace(" ",'') for i in x])
movies['cast']= movies['cast'].apply(lambda x:[i.replace(" ",'') for i in x])
```

```python
movies.head()
```

```python
movies.head()
```

```python
movies['tags']= movies['overview'] + movies['keywords'] + movies['genres'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
new_df= movies[['movie_id' ,'title','tags']]
```

```python
 new_df['tags']= new_df['tags'].apply(lambda x: " ".join(x) if isinstance(x, list) else "")
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]

    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags']= new_df['tags'].apply(stem)
```

```python
new_df['tags']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
k=cv.get_feature_names_out()
```

```python
print(k)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index=new_df[new_df['title'] == movie].index[0]
    distances=similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
